```python
import pandas as pd
import requests

req = requests.get('https://pastebin.com/raw/r1HwmQt0')
pokemon_dict = req.json()

pokemon_dict
```

```python
pokemon_names = []
pokemon_ids = []
for pokemon in pokemon_dict['Pokemon']:
    try:
        id_num, name = pokemon['Pokemon']['UniqueID'].split('_POKEMON_')
        pokemon_names.append(name)
        pokemon_ids.append(id_num)
    except: continue
```

```python
len(pokemon_ids)
```

```python
len(pokemon_names)
```

```python
df = pd.DataFrame(pokemon_names, index=pokemon_ids, columns=['name'])
```
