```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = []
for i in range(10000):
    res.append(np.random.binomial(1, .5, 100).sum())
    
sns.distplot(res, kde=False, bins=20)
```

```python
plt.hist(res, bins=20)
```

```python
res2 = []
for i in range(10000):
    res2.append(np.random.randint(1, 7, 30).sum())
    
sns.distplot(res2, kde=False, bins=20)
```

```python
res3 = []
for i in range(10000):
    res3.append(np.random.poisson(5, 5).sum())
    
sns.distplot(res3, kde=False, bins=20)
```
