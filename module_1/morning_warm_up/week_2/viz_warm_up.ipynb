```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("co2_mm_mlo.txt", sep = r"\s*", skiprows = 72, 
                 header = None, 
                 names = ['Year', 'Month', 'Inferred_Date', 
                          'CO2_average_ppm', 'Interpolated_CO2_ppm', 
                          'seasonal_trend',
                          'days'])

```

```python
co2 = df.iloc[:, [0, 1, 4]].copy()
```

```python
co2.head()
```

```python
co2_mean = co2.groupby(by='Year').mean()
```

```python
co2_months = co2.groupby(by='Month')['Interpolated_CO2_ppm']
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.plot(co2_mean['Interpolated_CO2_ppm'], c='r')
co2.plot(x='Year', y='Interpolated_CO2_ppm', ax=ax, kind='scatter', c='Month', colormap = 'BuPu', s=50)
ax.set_xlim(1977, 2020)
ax.set_ylim(300, 420)
ax.set_yticks(np.linspace(300, 420, 7))
ax.set_xticks(np.linspace(1980,2020,9))
ax.legend().remove()
ax.set
plt.show()
```
